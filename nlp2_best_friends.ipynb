```python
import math
import shutil
shutil.move('/content/NLP/nlp2_best_friends.py', '/content/nlp2_best_friends.py')
from nlp2_best_friends import *
def calculatePMI (fileIn,  distance, enc="iso8859_2"):
    lines = []
    for line in open(fileIn, "r", encoding=enc).readlines():
        lines.append(line.strip())
    limit = 20 #limit of output to file
    cBigr = countNgram(createBigrams(lines,distance))
    cUnigr = countNgram(createUnigrams(lines))
    pmi = PMI(cUnigк, cBigr, 10.0) #10 is limit of appearence

    count = 0
    item0 = []
    item1 = []

    for item in pmi:
        count += 1
        if count > limit: break
        item0.append(item[0])
        item1.append(item[1])
    return item0, item1
```
